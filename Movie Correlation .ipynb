```python
#Import Libraries 

import pandas as pd 
import seaborn as sns 
import numpy as np

import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
```

```python
mean_budget = df['budget'].mean()
df['budget'].replace(np.nan, mean_budget, inplace = True)
print('number of NaN values for column budget:',df['budget'].isnull().sum())
```

```python
mean_gross = df['gross'].mean()
df['gross'].replace(np.nan, mean_gross, inplace = True)
print('number of NaN values for column gross:',df['gross'].isnull().sum())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
#Change data type of columns 
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
#Sort by gross
df = df.sort_values(by=['gross'], inplace= False, ascending = False)
```

```python
#To View the full data 
pd.set_option('display.max_rows',None)
```

```python
#Drop duplicates

df.drop_duplicates()
```

```python
df.head()
```

```python
# Scatter plot for Budget vs Gross

plt.scatter(x=df['budget'], y= df['gross'])

plt.title('Budget vs Gross')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#Plot the budget vs groos using seaborn 

sns.regplot(x= df['budget'], y=df['gross'], data = df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Find the correlation 
df.corr()
```

## High correlation between budget and gross 

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation matrix for numerical atts')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Companies correlation 
df_normalize = df

for col_name in df_normalize.columns:
    if(df_normalize[col_name].dtype == 'object'):
        df_normalize[col_name] = df_normalize[col_name].astype('category')
        df_normalize[col_name] = df_normalize[col_name].cat.codes 
df_normalize        
```

```python
correlation_matrix = df_normalize.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation matrix for numerical atts')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_normalize.corr()
```

```python
correlation_mat = df_normalize.corr()

corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.4]
high_corr
```
